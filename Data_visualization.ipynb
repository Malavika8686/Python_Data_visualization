```python
'''
 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using
 a heatmap.
 '''
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
n_samples = 100
feature_1 = np.random.normal(0, 1, n_samples)
feature_2 = feature_1 + np.random.normal(0, 0.5, n_samples)
feature_3 = -feature_1 + np.random.normal(0, 0.5, n_samples)
feature_4 = np.random.normal(0, 1, n_samples)
df = pd.DataFrame({
    'Feature_1': feature_1,
    'Feature_2': feature_2,
    'Feature_3': feature_3,
    'Feature_4': feature_4
})

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='rainbow')
```

```python
'''
3. Using the sales data, generate a heatmap to visualize the variation in sales across
different months and days.
'''
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.express as px
data = {
 'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
 'Day': np.random.choice(range(1, 31), 100),
 'Sales': np.random.randint(1000, 5000, 100)
 }

df = pd.DataFrame(data)
pivot_table = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')
# Create the heatmap
fig = px.imshow(pivot_table,
                labels=dict(x="Month", y="Day", color="Average Sales"),
                title="Heatmap of Sales Variation Across Months and Days",
                color_continuous_scale='Viridis')
fig.show()
```

```python
'''
4. Using the given x and y data, generate a 3D surface plot to visualize the function z=sqrt(x^2+y^2)
'''
#code:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.express as px

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig = go.Figure(data=[go.Surface(z=df.pivot(index='X', columns='Y', values='Z').values,
                                 x=df.pivot(index='X', columns='Y', values='X').values,
                                 y=df.pivot(index='X', columns='Y', values='Y').values,
                                 colorscale='Viridis')])
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'),
                  autosize=False,
                  width=800,
                  height=800)

# Show the plot
fig.show()
```

```python
'''
                                            BOKEH ASSIGNMENT:
 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

'''
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show,output_file

import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer with x and y values
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Show the plot
show(p)
```

```python
'''
2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns
'''
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show,output_file

import numpy as np
import pandas as pd

# Generate random x and y values
np.random.seed(42)  # For reproducibility
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

# Generate random sizes and colors
sizes = np.random.uniform(5, 20, 100)
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], 100)

# Create a DataFrame for easier manipulation
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a new plot
p = figure(title="Scatter Plot with Variable Sizes and Colors", x_axis_label='x', y_axis_label='y', width=800, height=400)

# Add a scatter renderer with variable sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', legend_field='colors', source=data, fill_alpha=0.6)

# Show the plot
output_file("scatter_plot.html")
show(p)
```

```python
'''
 4. Create a Bokeh histogram to visualize the distribution of the given data.
'''
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show,output_file
import numpy as np

# Generate random data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)

# Add a quad renderer for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Show the plot
output_file("histogram.html")
show(p)
```
